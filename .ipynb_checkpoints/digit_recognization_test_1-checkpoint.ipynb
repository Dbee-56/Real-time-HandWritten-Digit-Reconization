# SCREEN CAPTURE

```python
import pyscreenshot as ImageGrab
import time
```

```python
images_folder='captured_images/9/'
for i in range(0,50):
    time.sleep(8)
    im=ImageGrab.grab(bbox=(60,250,550,700)) #x1,y1,x2,y2
    print("saved.....",i)
    im.save(images_folder+str(i)+'.png')
    print("clear screen and re draw again........")
```
